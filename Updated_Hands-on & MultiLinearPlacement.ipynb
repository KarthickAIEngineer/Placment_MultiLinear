```python
import pandas as pd
dataset=pd.read_csv("Placement.csv")
#Save univariate file and then execute the command or else will get import error
from univariate import Univariate
obj=Univariate()
quan,qual=obj.QuanQual(dataset)
```

```python
dataset.isnull().sum()
```

```python
dataset["salary"]=dataset["salary"].fillna(0)
```

```python
Uni=obj.uniAnalysis(dataset,quan)
Uni
```

```python
lesser=[]
greater=[]

for columnName in quan:
    if(Uni[columnName]["Lesser"]>Uni[columnName]["Min"]):
        lesser.append(columnName)
    if(Uni[columnName]["Greater"]<Uni[columnName]["Max"]):
        greater.append(columnName)
```

```python
greater
```

```python
for lesscolumn in lesser:
    dataset[lesscolumn][dataset[lesscolumn]<Uni[lesscolumn]["Lesser"]]=Uni[lesscolumn]["Lesser"]
for greatercolum in greater:
    dataset[greatercolum][dataset[greatercolum]>Uni[greatercolum]["Greater"]]=Uni[greatercolum]["Greater"]
```

```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = dataset
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
def stdNBgraph(dataset):
    #Coverted to standard Normal Distribution
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()

    values=[i for i in dataset]
    z_score=[((j-mean)/std) for j in values]
    sns.distplot(z_score,kde=True)
    sum(z_score)/len(z_score)
```

```python
#Multiple Linear Regression
#One Hot Encoding- conversion of string to number and dropping first column
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
#Input and output variables
indepVariable=dataset[['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p',
       'gender_M', 'ssc_b_Others', 'hsc_b_Others', 'hsc_s_Commerce',
       'hsc_s_Science', 'degree_t_Others', 'degree_t_Sci&Tech', 'workex_Yes',
       'specialisation_Mkt&HR', 'status_Placed']]
depVariable=dataset[['salary']]
```

```python
indepVariable
```

```python
depVariable
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(indepVariable,depVariable,test_size=0.30,random_state=0)

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
weight=regressor.coef_
bias=regressor.intercept_
print("Weight for Salary prediction {} ".format(weight)) #Weight=number of input parameters
print("Bias for Salary prediction {} ".format(bias))#one value
```

```python
#Evaluation phase for actual Vs predicted, displaying r2 value
y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
